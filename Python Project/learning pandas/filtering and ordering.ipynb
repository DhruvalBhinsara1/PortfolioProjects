# Filtering and Ordering in pandas

```python
import pandas as pd
```

```python
df = pd.read_csv(r'E:\PortfolioProjects\Python Project\learning pandas\world_population.csv')
```

```python
df
```

```python
df[df['Rank'] <= 10]
```

```python
specific_countrie = ['India', 'China', 'United States']
```

```python
df[df['Country'].isin(specific_countrie)]
```

```python
df[df['Country'].str.contains('United')]
```

```python
df[df['Country'].str.contains('In')]
```

```python
df2 = df.set_index('Country')
```

```python
df2.filter(items=['India'], axis=0)
```

```python
df2.filter(like='United', axis=0)
```

```python
df2.loc['United States']
```

```python
df2.iloc[3]
```

```python
df[df['Rank'] <= 10].sort_values(by=['Continent','Country'],ascending=[False,True])
```
