## What to Know  
- Pandas `.plot()` is a simple way to create lots of chart types without much code  
- Each kind of plot is better for showing certain patterns or comparisons  
- You can tweak things like colors, labels, and figure size to make graphs easier to read  
- Switching between plot types can highlight different insights from the same dataset  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = df.set_index("Date")
```

## Line Plot
I love line plots for tracking changes over time - they're my go-to when I want to see how things evolve. Looking at our ice cream ratings, you can really see the daily ups and downs in flavor, texture, and overall scores. It's fascinating how the ratings fluctuate and what that might tell us about the ice cream quality on different days.

```python
df.plot(kind='line', title='Ice Cream Rating', xlabel='Daily Rating', ylabel='Scores', figsize=(12, 6))
```

**Revision Notes:** `df.plot(kind='line')` creates line plots. Ideal for showing trends over time or ordered categories. Customize with `title`, `xlabel`, `ylabel`, `figsize`.

## Bar Plots
Bar charts are my favorite for making comparisons crystal clear. The stacked version shows you how different rating components add up for each day, while the regular bars let you compare individual values side by side. I find these super helpful when I want to quickly spot which days had the highest or lowest ratings.

```python
df.plot(kind='bar', title='Ice Cream Rating', xlabel='Daily Rating', ylabel='Scores', stacked=True, figsize=(12, 6))
```

```python
df.plot(kind='bar', title='Ice Cream Rating', xlabel='Daily Rating', ylabel='Scores', figsize=(12, 6))
```

**Revision Notes:** `df.plot(kind='bar')` for vertical bars, `df.plot.barh()` for horizontal. Use `stacked=True` to stack bars. Great for comparing categories.

## Horizontal Bar Plots
Sometimes I flip bars to horizontal when the labels are long or I want to emphasize ranking. These work great for our ice cream data - you can easily scan from top to bottom to see which days performed best. The stacked version gives you that extra layer of understanding about how the different ratings contribute to the total.

```python
df.plot.barh(stacked=True, figsize=(10, 6))
```

```python
df.plot.barh(figsize=(10, 6))
```

**Revision Notes:** Horizontal bars with `df.plot.barh()`. Useful when category names are long. `stacked=True` shows component contributions.

## Scatter Plots
When I want to see relationships between two continuous variables, scatter plots are my go-to. They're perfect for spotting correlations, clusters, or outliers in the data. For our ice cream ratings, this helps me understand if there's any pattern between different scoring aspects.

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', figsize=(8, 6))
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=150, c='lightblue', figsize=(8, 6))
```

**Revision Notes:** `df.plot.scatter(x='col1', y='col2')` for scatter plots. Shows relationships between two variables. Customize with `s` (size), `c` (color).

## Histograms
Histograms are fantastic for understanding the distribution of your data. They show me how values are spread out and help identify patterns like normal distributions, skewness, or multiple peaks. With our ice cream data, I can see how the ratings are distributed across different score ranges.

```python
df.plot.hist(bins=10, figsize=(10, 6))
```

```python
df.plot.hist(bins=100, figsize=(10, 6))
```

**Revision Notes:** `df.plot.hist(bins=10)` creates histograms. Shows data distribution. Adjust `bins` for detail level.

## Box Plot
Box plots are my favorite for quickly understanding the spread and outliers in data. They show the median, quartiles, and any extreme values at a glance. For our ice cream ratings, this helps me see the typical range of scores and identify any unusual days or ratings.

```python
df.boxplot(figsize=(10, 6))
```

**Revision Notes:** `df.boxplot()` shows quartiles, median, outliers. Great for understanding data spread and identifying anomalies.

## Area Plot
Area plots are great for showing cumulative totals over time or categories. They emphasize magnitude and can show how different components contribute to the whole. With stacked area plots, I can see both individual trends and the overall pattern in our ice cream data.

```python
df.plot.area(figsize=(12, 6), alpha=0.6)
```

**Revision Notes:** `df.plot.area()` creates area plots. Emphasizes magnitude and trends. Use `alpha` for transparency in overlaps.

## Pie Chart
Pie charts are perfect for showing proportions and percentages. They're best when you have a small number of categories and want to emphasize how each part contributes to the whole. For our ice cream data, this helps me visualize the distribution of ratings across different categories.

```python
# Use value counts or groupby to create meaningful pie slices; here we sum Flavor Rating by flavor name if available
if 'Flavor' in df.columns:
    pie_series = df.groupby('Flavor')['Flavor Rating'].sum()
else:
    pie_series = df['Flavor Rating']

# Plot pie with labels and improved size
pie_series.plot.pie(figsize=(7,7), autopct='%1.1f%%', startangle=140)
plt.ylabel('')
plt.title('Proportional Flavor Rating by Flavor')
```

**Revision Notes:** `df.plot.pie()` for pie charts. Best with few categories. Use `autopct` for percentages, `startangle` for rotation.

## My Visualization Tips

- I always adjust figure sizes based on what I'm trying to show - wider for time series, taller for comparisons
- The right chart depends on your question: trends need lines, comparisons need bars, relationships need scatter plots
- I avoid pie charts when I have too many small slices - a sorted bar chart usually works better
- Clear labels and titles save you from having to explain your charts later
- When styling gets in the way, I tweak figsize and transparency until everything looks right

## My Learnings

- Pandas `.plot()` is a powerful wrapper around matplotlib for quick visualizations
- Different plot types serve different purposes: trends (line), comparisons (bar), relationships (scatter), distributions (hist/box), compositions (pie/area)
- Always set appropriate `figsize` and labels for readability
- Style with `plt.style.use()` and customize colors/sizes as needed
- For complex data, preprocess (groupby, pivot) before plotting
- Remember: right chart type depends on the question you're asking
- Practice makes perfect - experiment with parameters to understand their effects
