# Web API's

```python
import requests 
import time
import random
import json

url = "https://api.citybik.es/v2/networks"
response = requests.get(url)
data = response.json()
networks = data['networks']
```

```python
print("Welcome to the City Bike Betting Game!")
print("You start with $1000.")
balance = 1000


#we could give option to see particular bike's details before place a bet,  such as its location, speed, and rental price. and will show random 5 bikes and the winning bike will be one of them.

while balance > 0:
    print(f"\nYour current balance is: ${balance}")
    bet_amount = int(input("Enter your bet amount: $"))
    
    if bet_amount > balance:
        print("You cannot bet more than your current balance.")
        continue
    
    print("Available bike networks:")
    for i, network in enumerate(networks[:5]):
        print(f"{i + 1}. {network['name']} ({network['location']['city']}, {network['location']['country']})")
    
    choice = int(input("Choose a network by entering its number: ")) - 1
    
    if choice < 0 or choice >= len(networks[:10]):
        print("Invalid choice. Please try again.")
        continue
    
    selected_network = networks[choice]
    print(f"You selected: {selected_network['name']} in {selected_network['location']['city']}")
    
    print("Placing your bet...")
    time.sleep(2)
    
    winning_network = random.choice(networks[:10])
    print(f"The winning network is: {winning_network['name']} in {winning_network['location']['city']}")
    
    if winning_network == selected_network:
        winnings = bet_amount * 2
        balance += winnings
        print(f"Congratulations! You won ${winnings}.")
    else:
        balance -= bet_amount
        print(f"Sorry, you lost your bet of ${bet_amount}.")
    
    if balance <= 0:
        print("You have run out of money. Game over!")
        break
    
    play_again = input("Do you want to play again? (yes/no): ").strip().lower()
    if play_again != 'yes':
        print(f"You leave the game with a balance of ${balance}. Thanks for playing!")
        break
```
