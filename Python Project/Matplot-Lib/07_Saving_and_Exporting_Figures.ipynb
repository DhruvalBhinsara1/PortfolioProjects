# Saving and Exporting Figures in Matplotlib

Matplotlib allows you to save plots in various formats for use in documents, presentations, or the web. In this section, we'll cover:
- Saving figures as PNG, PDF, SVG, etc.
- Adjusting resolution and size

## Saving as PNG
Use `plt.savefig()` to save the current figure.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('Save as PNG')
plt.savefig('plot.png')
plt.show()
```

## Saving as PDF or SVG
Change the file extension in `savefig()` to save in other formats.

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('Save as PDF')
plt.savefig('plot.pdf')
plt.title('Save as SVG')
plt.savefig('plot.svg')
plt.show()
```

## Adjusting Resolution and Size
Use the `dpi` and `figsize` arguments to control output quality and size.

```python
plt.figure(figsize=(6, 4), dpi=150)
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('High-Resolution Plot')
plt.savefig('high_res_plot.png', dpi=300)
plt.show()
```

## Summary
- Use `savefig()` to export plots in various formats.
- Adjust `dpi` and `figsize` for quality and size.
