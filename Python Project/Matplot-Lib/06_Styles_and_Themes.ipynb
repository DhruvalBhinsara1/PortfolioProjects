# Styles and Themes in Matplotlib

Matplotlib provides built-in styles and themes to quickly change the appearance of your plots. In this section, we'll cover:
- Listing available styles
- Applying styles
- Customizing styles

## Listing Available Styles
Use `plt.style.available` to see all built-in styles.

```python
import matplotlib.pyplot as plt
print(plt.style.available)
```

## Applying a Style
Use `plt.style.use()` to apply a style to your plots.

```python
plt.style.use('ggplot')
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('ggplot Style')
plt.show()
```

## Customizing Styles
You can further customize styles by changing parameters such as colors, line widths, and fonts.

```python
plt.style.use('seaborn-v0_8-dark')
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='purple', linewidth=3)
plt.title('Customized Style')
plt.show()
```

## Summary
- Use built-in styles for quick visual changes.
- Customize further with plot parameters.
