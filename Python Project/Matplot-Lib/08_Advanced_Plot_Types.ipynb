## Histogram
Histograms show the distribution of a dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Box Plot
Box plots show the distribution and outliers in a dataset.

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Box Plot')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.show()
```

## Pie Chart
Pie charts show proportions of a whole.

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```

## Heatmap
Heatmaps visualize matrix-like data with color coding.

```python
matrix = np.random.rand(5, 5)
plt.imshow(matrix, cmap='viridis', interpolation='nearest')
plt.title('Heatmap')
plt.colorbar()
plt.show()
```

## Summary
- Use histograms for distributions, box plots for spread, pie charts for proportions, and heatmaps for matrices.
