# Basic Plots in Matplotlib

Matplotlib supports a variety of basic plot types. In this section, we'll cover:
- Line Plots
- Scatter Plots
- Bar Plots

Each plot type will be demonstrated with code examples and explanations.

## Line Plot
A line plot is the most basic plot type in Matplotlib. It is used to display data points connected by straight lines.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## Scatter Plot
Scatter plots are used to display the relationship between two variables. Each point represents an observation.

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78, 77, 85, 86]

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## Bar Plot
Bar plots are used to compare quantities corresponding to different groups or categories.

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]

plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

## Summary
- Line plots show trends over intervals.
- Scatter plots show relationships between variables.
- Bar plots compare quantities across categories.
