## Task 1: Create a Line Plot for Temperature Data Across a Week

```python
import numpy as np
import matplotlib.pyplot as plt

# Mock temperature data for a week
days = np.array(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
temperatures = np.array([22, 24, 19, 23, 25, 20, 21])
x = np.arange(len(days))
plt.plot(x, temperatures, marker='o', linestyle='-', color='b')
plt.xticks(x, days)
plt.title('Temperature Data Across a Week')
plt.xlabel('Days of the Week')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

## Task 2: Plot a bar chart comparing marks of students in different subjects.

```python
import numpy as np
import matplotlib.pyplot as plt

students = ['Dhruval', 'Nishant', 'Suhail', 'Saurabh']
subjects = ['Math', 'Science', 'English']
marks = np.array([[85, 90, 78],
                  [88, 92, 80],
                  [90, 85, 82],
                  [75, 80, 70]])
x = np.arange(len(students))
width = 0.2
plt.bar(x - width, marks[:, 0], width, label='Math', color='r')
plt.bar(x, marks[:, 1], width, label='Science', color='g')
plt.bar(x + width, marks[:, 2], width, label='English', color='b')
plt.xticks(x, students)
plt.title('Students Marks in Different Subjects')
plt.legend()
plt.show()
```

## Task 3: display a random scatter plot showing height vs weight

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
heights = np.random.normal(170, 10, 100)
weights = np.random.normal(70, 15, 100)
plt.scatter(heights, weights, color='purple', alpha=0.6)
plt.title('Height vs Weight Scatter Plot')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

## Task 4: Create a pie chart for time spent on different acvtivities in a day.

```python
import numpy as np
import matplotlib.pyplot as plt
activities = ['Sleeping', 'Studying', 'Eating', 'Walking', 'Watching Anime']

time_spent = [8, 6, 2, 3, 5]
colors = ["#e28585",'#66b3ff','#99ff99','#ffcc99', '#c2c2f0']
plt.pie(time_spent, labels=activities, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Time Spent on Different Activities in a Day')
plt.axis('equal')
plt.show()
```

## Task 5: Use a subplot to compare different plot types for the same data.

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(data, color='blue')
plt.title('Line Plot')
plt.subplot(2, 2, 2)
plt.hist(data, bins=30, color='green', alpha=0.7)
plt.title('Histogram')
plt.subplot(2, 2, 3)
plt.scatter(np.arange(len(data)), data, color='red', alpha=0.5)
plt.title('Scatter Plot')
plt.subplot(2, 2, 4)
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Box Plot')
plt.tight_layout()
plt.show()
```

## Task 6: Load a dataset using Pandas and visualize column statistics using Matplotlib.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'Salary': [50000, 60000, 55000, 70000, 65000]
}
df = pd.DataFrame(data)
df.to_csv('people_data.csv', index=False)
df = pd.read_csv('people_data.csv')
plt.figure(figsize=(10, 5))
plt.bar(df['Name'], df['Salary'], color='skyblue')
plt.title('Salary of Individuals')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.show()
```
