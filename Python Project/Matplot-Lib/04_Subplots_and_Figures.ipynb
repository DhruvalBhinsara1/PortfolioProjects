# Subplots and Figures in Matplotlib

Matplotlib allows you to create multiple plots in a single figure using subplots. This is useful for comparing different datasets or visualizations side by side.

In this section, we'll cover:
- Creating subplots
- Different subplot layouts
- Adjusting spacing between subplots

## Creating Subplots with `plt.subplot`
The `subplot` function allows you to create a grid of plots within a single figure.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))

#Graph 1
#plt.sublplot(value,value,value) what each value do?


plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title('Plot 1')


#Graph 2
plt.subplot(1, 2, 2)
plt.plot([1, 2, 3], [1, 2, 3])
plt.title('Plot 2')

plt.tight_layout()
plt.show()
```

## Using `plt.subplots` for Flexible Layouts
The `subplots` function provides more control and returns figure and axes objects.

```python
fig, axs = plt.subplots(2,2 ,figsize=(8, 6))
axs[0, 0].plot([1, 2, 3], [1, 4, 9])
axs[0, 0].set_title('Top Left')
axs[0, 1].bar(['A', 'B', 'C'], [5, 7, 3])
axs[0, 1].set_title('Top Right')
axs[1, 0].scatter([1, 2, 3], [3, 2, 1])
axs[1, 0].set_title('Bottom Left')
axs[1, 1].hist([1, 2, 2, 3, 3, 3], bins=3)
axs[1, 1].set_title('Bottom Right')
plt.tight_layout()
plt.show()
```

## Adjusting Spacing
Use `plt.tight_layout()` to automatically adjust subplot spacing for better appearance.

## Summary
- Use `subplot` or `subplots` for multiple plots in one figure.
- Adjust layout with `tight_layout()`.
